```python
import datetime
from collections import defaultdict
```

```python
class Event:
    def __init__(self, date, event_type, user, machine):
        self.date = date
        self.event_type = event_type
        self.user = user
        self.machine = machine
```

```python
login_count = 0
logout_count = 0
machine_usage_count = defaultdict(int)
login_members = []
logout_members = []
machines_used = []
```

```python
def log_event(event):
    global login_count, logout_count, login_members, logout_members, machines_used
    with open("usage_log.txt", "a") as log_file:
        if event.event_type == "login":
            log_file.write(f"Date and Time: {event.date} User type: {event.event_type} User Name: {event.user} Machine Name: {event.machine}\n")
            login_count += 1
            machine_usage_count[event.machine] += 1
            login_members.append(event.user)
            machines_used.append(event.machine)
        else:
            log_file.write(f"Date and Time: {event.date} User type: {event.event_type} User Name: {event.user}\n")
            logout_count += 1
            logout_members.append(event.user)
```

```python
def login(user):
    current_time = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    machine_used = input('Enter which machine you used: ')
    login_event = Event(current_time, "login", user, machine_used)
    log_event(login_event)
    print(f"User {user} logged in at {current_time} and used {machine_used}")
```

```python
def logout(user):
    current_time = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    logout_event = Event(current_time, "logout", user, "")
    log_event(logout_event)
    print(f"User {user} logged out at {current_time}")
```

```python
def print_summary():
    global login_count, logout_count, machine_usage_count, login_members, logout_members, machines_used
    print('############################')
    print(f"Login count: {login_count}")
    print(f"Logout count: {logout_count}")
    print("\nMachine Usage Summary:")
    for machine, count in machine_usage_count.items():
        print(f"Machine: {machine} - Usage Count: {count}")

    print("\nLogin Members:")
    print(login_members)

    print("\nLogout Members:")
    print(logout_members)

    print("\nMachines Used:")
    print(machines_used)

    print('#############################')
    with open("usage_summary.txt", "a") as summary_file:
        summary_file.write(f"##############################\n")
        summary_file.write(f"Login count: {login_count}\n")
        summary_file.write(f"Logout count: {logout_count}\n")
        summary_file.write(f"\nMachine Usage Summary:\n")
        for machine, count in machine_usage_count.items():
            summary_file.write(f"Machine: {machine} - Usage Count: {count}\n")

        summary_file.write("\nLogin Members:\n")
        summary_file.write(str(login_members) + '\n')

        summary_file.write("\nLogout Members:\n")
        summary_file.write(str(logout_members) + '\n')

        summary_file.write("\nMachines Used:\n")
        summary_file.write(str(machines_used) + '\n')

        summary_file.write(f"##############################\n")
```

```python
# Example usage:
while True:
    print("Choose an action:")
    print("1. Login")
    print("2. Logout")
    print("3. Print Summary")
    print("4. Exit")

    choice = input("Enter your choice (1/2/3/4): ")

    if choice == "1":
        username = input("Enter the username for login: ")
        login(username)
    elif choice == "2":
        username = input("Enter the username for logout: ")
        logout(username)
    elif choice == "3":
        print_summary()
    elif choice == "4":
        break
    else:
        print("Invalid choice. Please enter 1, 2, 3, or 4.")
```
